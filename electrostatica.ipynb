  
# Campo eléctrico generado por dos cargas fuente
  
  Recordemos la expresión del campo eléctrico sobre el **punto de observación** $\vec{r}$:

\begin{equation}
\vec{E}(\vec{r})=k\sum_k^N\frac{q_k^\prime(\vec{r}-\vec{r}_k^\prime)}{\mathopen|\vec{r}-\vec{r}_k^\prime\mathclose|^3}
\end{equation}

Según esta expresión, se debe realizar una sumatoria sobre las **cargas fuentes** $q_k^\prime$, las cuales se ubican en las posiciones $\vec{r}_k^\prime$.

Si se desea conocer la fuerza ejercida por este campo eléctrico sobre una **carga de prueba** $q$ en el punto de observación $\vec{r}$, hay que multipicar el campo eléctrico por el valor de la carga de prueba

\begin{equation}
\vec{F}(\vec{r})=q\vec{E}(\vec{r})=kq\sum_k^N\frac{q_k^\prime(\vec{r}-\vec{r}_k^\prime)}{\mathopen|\vec{r}-\vec{r}_k^\prime\mathclose|^3}
\end{equation}

La propuesta de este notebook es crear un código que calcule el campo  (en 2D) y la fuerza sobre una carga debido a dos (2) cargas fuente.


# Instalamos los paquetes necesarios y definimos constantes

```python
import numpy as np
import matplotlib.pyplot as plt

k = 8.9875e9  # Constante de Coulomb en N*m^2/C^2
```

# Definimos el punto de observación y el valor de la carga de prueba

```python
r = np.array([0,1]) # Punto de observación [m]
q = 2e-6  # Carga de prueba [C]
```

# Definimos las posiciones y los valores de las cargas de prueba

```python
r1 = np.array([-1,0]) # Posición de la carga fuente 1 [m]
r2 = np.array([1,0]) # Posición de la carga fuente 2 [m]
q1 = 1e-6  # Carga fuente 1 [C]
q2 = -1e-6  # Carga fuente 2 [C]
```

# Calculamos el campo eléctrico y la fuerza en $\vec{r}$

```python
E1 = k*q1*(r-r1)/((r[0]-r1[0])**2 + (r[1]-r1[1])**2)**(3/2)
E2 = k*q2*(r-r2)/((r[0]-r2[0])**2 + (r[1]-r2[1])**2)**(3/2)
E = E1 + E2

F1 = q*E1
F2 = q*E2
F = q*E
```

# Grafico vectores de campo eléctrico

```python
# Grafico el campo eléctrico
escala = np.max([np.abs(E[0]), np.abs(E[1])])

plt.quiver(r[0], r[1], E1[0], E1[1], label=r'$E_1$', angles='xy', scale_units='xy',scale=escala,  color = 'r' )
plt.quiver(r[0], r[1], E2[0], E2[1], label=r'$E_1$', angles='xy', scale_units='xy',scale=escala,  color = 'b')
plt.quiver(r[0], r[1], E[0], E[1], label =r'$E$', angles='xy', scale_units='xy',scale=escala)

# Dibujo cargas fuente
plt.scatter(r1[0], r1[1], c='r', s=50, marker='o', label = 'q1 = ' + str(q1) + ' C')
plt.scatter(r2[0], r2[1], c='b', s=50, marker='o', label = 'q2 = ' + str(q2) + ' C')

# Dibujo carga de prueba
plt.scatter(r[0], r[1], c='g', s=50, marker='o', label = 'q = ' + str(q) + ' C')
plt.legend()
plt.title('Campos eléctricos sobre la carga $q$')
plt.ylim(-2, 2)
plt.xlim(-2, 2)
plt.grid()
plt.show()

```

# Gráficos de fuerzas

```python
# Grafico la fuerzas
escala = np.max([np.abs(F[0]), np.abs(F[1])])
plt.quiver(r[0], r[1], F1[0], F1[1], label=r'$F_1$', angles='xy', scale_units='xy',scale=escala,  color = 'r' )
plt.quiver(r[0], r[1], F2[0], F2[1], label=r'$F_2$', angles='xy', scale_units='xy',scale=escala,  color = 'b')
plt.quiver(r[0], r[1], F[0], F[1], label =r'$F$', angles='xy', scale_units='xy',scale=escala)

# Dibujo cargas fuente
plt.scatter(r1[0], r1[1], c='r', s=50, marker='o', label = 'q1 = ' + str(q1) + ' C')
plt.scatter(r2[0], r2[1], c='b', s=50, marker='o', label = 'q2 = ' + str(q2) + ' C')

# Dibujo carga de prueba
plt.scatter(r[0], r[1], c='g', s=50, marker='o', label = 'q = ' + str(q) + ' C')
plt.legend()
plt.title('Fuerzas sobre la carga $q$')
plt.ylim(-4, 4)
plt.xlim(-4, 4)
plt.grid()
plt.show()
```

# Campo eléctrico generado por una corona circular de radios $a$ y $b$; y densidad $\sigma$.

En clase se resolvió el siguiente ejercicio que consistía en calcular el campo eléctrico sobre el eje de simetría de una corono de radios  $a$ y $b$ y densidad $\sigma$.
Se llegó al siguiente resultado para el campo eléctrico sobre el eje de simetría era:

$$ \vec{E}(0,0,z)=\frac{\sigma z}{2\epsilon_0}\left[  \frac{1}{\sqrt{a^2 + z^2}} -  \frac{1}{\sqrt{b^2 + z^2}} \right]\hat{z}$$

A continuación se grafica el resultado para algunos casos particulares.




```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.constants import epsilon_0

# Definimos constantes y parámetros del problema

k = 1/(4*np.pi*epsilon_0) # k: Constante de Coulomb [N·m^2/C^2]
e_0 = 8.854e-12 # Constante de interacción electrostática [F/m]
a = 1 # Radio interno de la corona [m]
b = 2 # Radio externo de la corona [m]
sigma = 1e-6 # Densidad de carga de la corona [C/m^2]

# Genero vector de z
z = np.linspace(-10, 10, 1000) # [m]

# Calculo campo electrico
Ez = sigma*z/2/e_0*(1/np.sqrt(a**2 + z**2) - 1/np.sqrt(b**2 + z**2))


# Plotting multiple lines on a single plot
plt.plot(z, Ez, label= 'a = ' + str(a) + '; b = ' + str(b), color='b')

#Ejes y título
plt.ylabel('Ez')
plt.xlabel('z')

plt.title('Campo eléctrico sobre el eje de una corona.')
plt.grid()
plt.legend()
plt.show()
```

Notemos que en el centro del disco, el campo es nulo. Luego comienza a crecer y finalmente adopta la forma $\frac{1}{r^2}$ (lo vemos más adelante)

# Campo eléctrico de un disco carga con densidad $\sigma$

Para calcular el campo de un disco, se puede tomar el caso particular $a=0$ del ejercico anterior. Con eso se obtiene:

```python
a = 0 # Radio interno de la corona [m]
b = 2 # Radio externo de la corona [m]
sigma = 1e-6 # Densidad de carga de la corona [C/m^2]

# Genero vector de z
z = np.linspace(-10, 10, 1000) # [m]

# Calculo campo electrico
Ez = sigma*z/2/e_0*(1/np.sqrt(a**2 + z**2) - 1/np.sqrt(b**2 + z**2))


# Plotting multiple lines on a single plot
plt.plot(z, Ez, label= 'a = ' + str(a) + '; b = ' + str(b), color='b')

#Ejes y título
plt.ylabel('Ez')
plt.xlabel('z')

plt.title('Campo eléctrico sobre el eje de un disco.')
plt.grid()
plt.legend()
plt.show()
```

Notar que ahora aparece una discontinuidad del campo en el plano del disco. Se puede ver del resultado que este salto corresponde a $\sigma/\epsilon_0$. Este resultado es general, vale para el salto del campo en cualquier superficie cargada.

# Límite de radio infinito (plano cargado).

Otro caso para analizar es el límite $b→∞$, con $a=0$. En este caso, se puede ver que la expresión del campo adopta la forma $$ \vec{E}(0,0,z)=\frac{\sigma }{2\epsilon_0}\left[  \frac{z}{|z|} \right]\hat{z}$$

Si bien parece existir una dependencia en $z$, en realidad se trata sólo de la función $signo(z)$.

```python
a = 0 # Radio interno de la corona [m]
b = 1 # Radio externo de la corona [m]
sigma = 1e-6 # Densidad de carga de la corona [C/m^2]

# Genero vector de z
z = np.linspace(-1, 1, 1000) # [m]

# Calculo campo electrico
Ez = sigma*z/2/e_0*(1/np.sqrt(a**2 + z**2) - 1/np.sqrt(b**2 + z**2))
E_inf = sigma/2/e_0*np.sign(z)

# Plotting multiple lines on a single plot
plt.plot(z, Ez, label= 'a = ' + str(a) + '; b = ' + str(b), color='b')
plt.plot(z, E_inf, label= 'plano infinito', color='r')



#Ejes y título
plt.ylabel('Ez')
plt.xlabel('z')

plt.title('Campo eléctrico sobre el eje de un disco.')
plt.grid()
plt.legend()
plt.show()
```

# Aproximación por carga puntual

Si se vuelve al resultado inicial, de una corona, vemos que la dependencia es particular. Sin embargo, veremos que a una distancia lo suficientemente grande (mucho mayor que el tamaño de la distribución de las cargas fuente), el campo eléctrico puede aproximarse al de una carga puntual en el centro de la corona.

La carga total está dada por $$ q = \sigma \pi \left(  b^2 - a^2 \right)$$

Comparemos ahora las expresiones $\frac{\sigma z}{2\epsilon_0}\left[  \frac{1}{\sqrt{a^2 + z^2}} -  \frac{1}{\sqrt{b^2 + z^2}} \right]$ vs  $\frac{q}{4\pi\epsilon_0}\frac{1}{z^2} $

```python
k = 1/(4*np.pi*epsilon_0) # k: Constante de Coulomb [N·m^2/C^2]
e_0 = 8.854e-12 # Constante de interacción electrostática [F/m]
a = 1 # Radio interno de la corona [m]
b = 2 # Radio externo de la corona [m]
sigma = 1e-6 # Densidad de carga de la corona [C/m^2]

# Calculo la carga total
q = sigma*np.pi*(b**2 - a**2)

# Genero vector de z
z = np.linspace(-10, 10, 100) # [m]

# Calculo campo electrico
Ez = sigma*z/2/e_0*(1/np.sqrt(a**2 + z**2) - 1/np.sqrt(b**2 + z**2)) # Corona
E = q/(4*np.pi*e_0*z**2)*np.sign(z)


# Plotting multiple lines on a single plot
plt.plot(z, Ez, label= 'a = ' + str(a) + '; b = ' + str(b), color='b')
plt.plot(z, E, label= 'puntual', color='r')

#Ejes y título
plt.ylabel('Ez')
plt.xlabel('z')
plt.ylim([-15000,15000])


plt.title('Campo eléctrico sobre el eje de una corona.')
plt.grid()
plt.legend()
plt.show()
```

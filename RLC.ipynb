# Script para graficar el módulo y la fase de la corriente en un circuito RLC serie:

```python
# Importo librerias numpy y matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defino valores de componentes y fuente de tension

V0 = 1    # Amplitud de la tensión, en volt
R = 10   # Resistencia, en ohm (1 kohm es lo que hay )
C = 1e-7  # Capacidad, en farad (100 nanoF es lo que hay)
L = 1e-5  # Inductancia, en henry

w0 = 1/np.sqrt(L*C)
print('La frecuencia angular de resonancia es ' + str(int(w0)) + ' rad/s.')
```

```python
# Defino vector de frecuencias angulares
w = np.logspace(1, 8, num=1000)

# Defino la impedancia del circuito
Z = R + 1j*(w*L-1/(w*C))

# Calculo la corriente compleja
I = V0/Z
```

```python
# Realizo figura de amplitud
plt.semilogx(w, np.abs(I))
plt.semilogx([1000,10000],[0.001,0.002],'k*') # Ejemplo de dos mediciones
plt.xlabel("Frecuencia angular, $\omega$ [rad/s]")
plt.ylabel("Corriente [A]")
plt.grid()
plt.show()
```

```python
# Realizo figura de fase de la corriente
plt.semilogx(w, np.angle(I)*180/np.pi)
plt.semilogx([1000,1000000],[90,0],'ks') # Ejemplos de mediciones
plt.xlabel("Frecuencia angular, $\omega$ [rad/s]")
plt.ylabel("Fase de la corriente [grados]")
plt.grid()
```

```python
# Realizo figura de Potencia
plt.semilogx(w, 0.5*V0*V0/R*np.cos(np.angle(Z)))
plt.xlabel("Frecuencia angular, $\omega$ [rad/s]")
plt.ylabel("Potencia [W]")
plt.grid()
```

```python
# Calculo el ancho de la curva de potencia
Dw = R/L
print('El ancho de la curva de potencia es ' + str(int(Dw)) + ' Hz')
# Parámetro Q
Q = w0*L/R
print('El parámetro Q del circuito es ' + str(int(Q)))
```

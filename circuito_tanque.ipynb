# Oscilador LC (circuito tanque)
Se denomina *circuito tanque* a la descarga de un capacitor sobre un inductor. Consideremos la siguiente situación, donde un capacitor de capacidad C es cargado inicialmente con una tensión V (llave en posición A). Luego se conmuta la llave a la posición B, donde el capacitor es descargado sobre un inductor de inductancia L. La situación se esquematiza en la siguiente figura:

image.png

(imagen de https://www.electronics-tutorials.ws/oscillator/oscillators.html)

Debemos considerar que, luego de permutada la llave, el inductor y el capacitor se encuentran en paralelo. Por este motivo, la tensión sobre los dos componentes debe ser la misma: $$\frac{Q}{C}=L\frac{di}{dt}$$
Recordemos que utilizamos la *convención de signos pasiva*. Esto significa que en el caso de un inductor en el cual la corriente **entra** por el terminal positivo, caída de potencial viene dada por $V_L= L\frac{di}{dt} $.
Siguiendo la misma convención, la corriente que **sale** de un capacitor por el terminal positivo viene dada por $i_C=-\frac{dQ}{dt}$. Notando que la corriente sobre el inductor es la misma que la corriente sobre el capacitor, se llega entonces a la expresión que rige el comportamiento de la carga en el capacitor:

$$\frac{Q}{C}= - L \frac{d^2Q}{dt^2} $$

Como se vio en clase, si se considera que la descarga sucede en $t=0$, la expresión para la carga del capacitor está dada por:
$$Q(t) = Q_0\cos(\omega t)$$
Con $\omega = \frac{1}{\sqrt{LC}}$ la frecuencia angular de la oscilación y $Q_0$ la carga incial del capacitor: $Q_0 = CV$.

Luego la corriente del circuito está dada por $$I(t)=Q_0\omega\sin(\omega t)=\frac{CV}{\sqrt{LC}}\sin(\omega t)=I_0\sin(\omega t)$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos valores de capacidad, inductancia y tensión de la fuente

C = 1e-6 # Capacidad en farad
L = 3e-3 # Inductancia en henry
V = 2 # Tensión volt,

# Carga inicial del capacitor
Q_0 = C*V

# Frecuencia angular de oscilación
omega = 1/np.sqrt(L*C)
```

```python
# Generamos un rango de valores de tiempo desde 0 hasta 4π (dos períodos de oscilación)

t = np.linspace(0, 4 * np.pi /omega, 100)

# Calcular el seno de cada valor de x
Q = Q_0*np.cos(omega * t)

# Crear la figura y el eje
plt.figure(figsize=(8, 6))

# Trazar la función seno
plt.plot(t, Q, label='Q(t)', color='blue', linewidth=2)

# Añadir etiquetas y título
plt.xlabel('t [segundos]', fontsize=14)
plt.ylabel('Q(t) [coulomb]', fontsize=14)
plt.title('Carga en el capacitor', fontsize=16)

# Añadir una cuadrícula
plt.grid(True)

# Añadir una leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```

```python
# Graficamos la corriente del circuito

I = omega*Q_0*np.sin(omega*t)

# Crear la figura y el eje
plt.figure(figsize=(8, 6))

# Trazar la función seno
plt.plot(t, I, label='I(t)', color='red', linewidth=2)

# Añadir etiquetas y título
plt.xlabel('t [segundos]', fontsize=14)
plt.ylabel('I(t) [ampere]', fontsize=14)
plt.title('Corriente en el circuito LC', fontsize=16)

# Añadir una cuadrícula
plt.grid(True)

# Añadir una leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```

Las energías en el capacitor e inductor se definen como: $$U_C = \frac{1}{2C}Q(t)² $$ y $$ U_L = \frac{L}{2}I(t)^2$$
Como se vio en clase, la energia oscila entre los dos componentes pero la suma permanece constante. También notamos que la expresión para las energías no son funciones armónicas y que su período equivale a la mitad del período $T = \frac{2\pi}{\omega}$.

```python
# Calculamos las energías en el inductor y capacitor

U_C = 0.5/C*np.square(Q)
U_L = 0.5*L*np.square(I)

# Crear la figura y el eje
plt.figure(figsize=(8, 6))

# Trazar la función seno
plt.plot(t, U_C, label='Uc(t)', color='red', linewidth=2)
plt.plot(t, U_L, label='UL(t)', color='blue', linewidth=2)


# Añadir etiquetas y título
plt.xlabel('t [segundos]', fontsize=14)
plt.ylabel('Uc', fontsize=14)
plt.title('Energía en circuito LC', fontsize=16)

# Añadir una cuadrícula
plt.grid(True)

# Añadir una leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```
